```python
url="D:\\MASAI\\Order food online from India's best food delivery service. Order from restaurants near you.html"
```

```python
headers={
    'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/109.0.0.0 Safari/537.36 OPR/95.0.0.0'

}
```

```python
hrefs = []
name_res = []
cuisine_res =[]
price_for_two_res = []
```

```python
with open(url, "r") as f:
    site = f.read()
soup = BeautifulSoup(site, "html.parser")
link_object = soup.findAll("a","_1j_Yo")
names = soup.findAll("div","nA6kb")
cuisines = soup.findAll("div","_1gURR")
price_4_two = soup.findAll("div", class_="nVWSi")
```

```python
for link in link_object:
    hrefs.append(link["href"])
```

```python
len(hrefs)
```

```python
for name in names:
    name_res.append(name.text)
```

```python
for c in cuisines:
    cuisine_res.append(c.text)
```

```python
for p in price_4_two:
    price_for_two_res.append(p.get_text())
```

```python
len(price_for_two_res)
```

```python
data = {'href': hrefs, 'name': name_res, 'cuisine': cuisine_res, 'price_4_two': price_for_two_res}
df = pd.DataFrame(data)
```

# Part 2

```python
store_url=[]
rating_res=[]
dishes_price=[]
dishes=[]
```

```python
for url in hrefs:
    response = requests.get(url, headers=headers)
    
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')
        
        ratings = soup.findAll("span","RestaurantRatings_totalRatings__3d6Zc")
        for i in ratings:
            rating_res.append(i.text)
```

```python
len(rating_res)
```

```python
print(len(store_url))
```

```python
print(len(rating_res))
```

```python
for url in hrefs:
    response = requests.get(url, headers=headers)
    
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')
        dish_name = soup.findAll("h3","styles_itemNameText__3ZmZZ")
        for i in dish_name:
            dishes.append(i.text)
```

```python
for url in hrefs:
    response = requests.get(url, headers=headers)
    
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')
        dish_price = soup.findAll("span","rupee")
        for i in dish_price:
            dishes_price.append(i.text)
```

```python
max_len = max(len(store_url), len(rating_res), len(dishes), len(dishes_price))
```

```python
data = {'store_url': store_url,
        'rating': rating_res + [np.nan]*(max_len-len(rating_res)),
        'dish_name': dishes + [np.nan]*(max_len-len(dishes)),
        'dish_price': dishes_price + [np.nan]*(max_len-len(dishes_price))}
```

```python
pd.DataFrame(data).to_csv("swiggy_table2.csv", index=False)
```
